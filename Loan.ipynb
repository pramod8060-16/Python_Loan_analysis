```python
'''1.What is the average current loan amount for 
each loan status?'''
df.groupby(by='Loan Status')['Current Loan Amount'].mean()
```

```python
'''2.How does the credit score vary with the annual 
income? '''
df.groupby(by='Credit Score')['Annual Income'].sum()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual data if not already loaded as df)
# df = pd.DataFrame(...)

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df["Annual Income"], df["Credit Score"], alpha=0.5, color="blue")
plt.title("Relationship Between Annual Income and Credit Score")
plt.xlabel("Annual Income")
plt.ylabel("Credit Score")
plt.grid(True)
plt.show()

# Optional: Analyze trends by income ranges
income_bins = pd.cut(df["Annual Income"], bins=5)  # Divide annual income into 5 bins
average_credit_score_by_income = df.groupby(income_bins)["Credit Score"].mean().reset_index()

print("Average Credit Score by Income Range:")
print(average_credit_score_by_income)
```

```python
'''3.Is there a correlation between the number of 
open accounts and the current credit balance? '''
import seaborn as sns
sns.scatterplot(data=df,x='Current Credit Balance',y='Number of Open Accounts')
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual data if not already loaded as df)
# df = pd.DataFrame(...)

# Calculate the correlation coefficient
correlation = df["Number of Open Accounts"].corr(df["Current Credit Balance"])
print(f"Correlation between Number of Open Accounts and Current Credit Balance: {correlation:.2f}")

# Scatter plot to visualize the relationship
plt.figure(figsize=(10, 6))
plt.scatter(df["Number of Open Accounts"], df["Current Credit Balance"], alpha=0.5, color="purple")
plt.title("Relationship Between Number of Open Accounts and Current Credit Balance")
plt.xlabel("Number of Open Accounts")
plt.ylabel("Current Credit Balance")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Sample DataFrame (replace this with your actual data if not already loaded as df)
# df = pd.DataFrame(...)

# Box plot to show the distribution of annual income by loan status
plt.figure(figsize=(12, 6))
df.boxplot(column="Annual Income", by="Loan Status", grid=False)
plt.title("Distribution of Annual Income Across Loan Statuses")
plt.suptitle("")  # Remove the automatic title
plt.xlabel("Loan Status")
plt.ylabel("Annual Income")
plt.xticks(rotation=45)
plt.show()

# Optional: You could also use histograms to show the frequency distribution for each loan status
plt.figure(figsize=(12, 6))
df[df["Loan Status"] == "Approved"]["Annual Income"].hist(alpha=0.5, label="Approved", bins=30, color="blue")
df[df["Loan Status"] == "Denied"]["Annual Income"].hist(alpha=0.5, label="Denied", bins=30, color="red")
plt.title("Histogram of Annual Income Across Loan Statuses")
plt.xlabel("Annual Income")
plt.ylabel("Frequency")
plt.legend()
plt.show()
```
